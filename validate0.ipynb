# compare/diff branchwater `manysearch` vs `sourmash search --containment`

They _should_ return identical results!

repo: https://github.com/sourmash-bio/2023-branchwater-validate/

```python
ls outputs/
```

```python
import glob
import pandas as pd

dfs = []
for cont in glob.glob("outputs/*.cont.csv"):
    df = pd.read_csv(cont)
    dfs.append(df)
    
cont_df = pd.concat(dfs)
cont_df.rename(columns={'similarity': 'containment',
                       'name': 'match_name'}, inplace=True)
cont_df[['match_name', 'containment', 'query_name']]
```

```python
cont_df = cont_df[cont_df.containment > 0.1]
cont_df2 = cont_df[['match_name', 'containment', 'query_name']]
cont_df2
```

```python
manysearch_df = pd.read_csv('outputs/manysearch.csv')
manysearch_df = manysearch_df[manysearch_df.containment > 0.1]
manysearch_df2 = manysearch_df[['match_name', 'containment', 'query_name']]
manysearch_df2
```

```python
branchwater_set = set()
for metag, cont, mag in manysearch_df2.itertuples(index=False):
    cont = round(cont, 5)
    branchwater_set.add((metag, cont, mag))

branchwater_set
```

```python
sourmash_set = set()
for metag, cont, mag in cont_df2.itertuples(index=False):
    cont = round(cont, 5)
    sourmash_set.add((metag, cont, mag))

sourmash_set
```

```python
branchwater_set - sourmash_set
```

```python
sourmash_set - branchwater_set
```

```python
assert sourmash_set == branchwater_set
```
